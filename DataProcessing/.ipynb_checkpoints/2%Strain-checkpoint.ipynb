# Temperature change

```python
TC = pd.read_excel('../../Single_reciprocating/0201-4p_12%_5.9L4H5.5U4C_Openloop.xlsx')
TC.tail(150)
```

```python
fig,ax = plt.subplots(figsize=(12,8))

tube4 = ax.plot(TC['time'][3:], (TC['1_TC02'][3:])-(TC['0_TC01'][3:]), color = 'darkorange')

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
ax.set_xlabel('Time [min]', fontsize = 18)
ax.set_ylabel('Delta_T [K]', fontsize = 18)
ax.set_title('Average strain = 4% \n HTP = 12%', fontsize = 20)
ax.set_xlim(0,35)
#ax.set_ylim(14,16)
#ax.legend(fontsize = 18)
ax.grid()
```

## Temperature change of individual tube

```python
fig,ax = plt.subplots(figsize=(12,8))

tube4 = ax.plot(TC['time'][3:]-2, (TC['0_TC01'][3:]-2)-(TC['1_TC02'][3:]-2), color = 'darkorange', label = 'tube4')
tube3 = ax.plot(TC['time'][3:]-2, (TC['2_TC03'][3:]-2)-(TC['3_TC04'][3:]-2), color = 'thistle', label = 'tube3')
tube2 = ax.plot(TC['time'][3:]-2, (TC['4_TC05'][3:]-2)-(TC['5_TC06'][3:]-2), color = 'yellowgreen', label = 'tube2')
tube1 = ax.plot(TC['time'][3:]-2, (TC['6_TC07'][3:]-2)-(TC['7_TC08'][3:]-2), color = '#6495ED', label = 'tube1')

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
ax.set_xlabel('Time [min]', fontsize = 18)
ax.set_ylabel('Delta_T [K]', fontsize = 18)
ax.set_title('4 bundle open loop \n Average strain = 3.0% \n HTP=16%', fontsize = 20)
ax.set_xlim(0,40)
ax.legend(fontsize = 18)
ax.grid()
```

## Temperature lift

```python
fig,ax = plt.subplots(figsize=(8,10))

th = ax.plot(TC['time'][3:], TC['1_TC02'][3:], color = 'darkorange', label = 'T$_h$')
tc = ax.plot(TC['time'][3:], TC['0_TC01'][3:], color = '#6495ED', label = 'T$_c$')
t_tank = ax.plot(TC['time'][3:], TC['10_TC11'][3:], color = 'tan', label = 'T$_{tank}$')

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
ax.set_xlabel('Time [min]', fontsize = 18)
ax.set_ylabel('Temperature [K]', fontsize = 18)
# ax.text(0, 5, 'Parabola $Y = x^2$', fontsize = 22)
ax.set_title('Temperature lift \n Average strain = 4% \n HTP=12%', fontsize = 20)
ax.set_xlim(0,35)
#ax.set_ylim(-20,20)
ax.legend(fontsize = 18)
ax.grid()
```
